```python
# Importing Libraries
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
```

```python
cars_data = pd.read_csv("used_car_dataset.csv")
print(cars_data)
```

```python
#1 DATA CLEANING
```

```python
# Deleting Extra Rows

filter = cars_data.loc[5000:]
cars_data.drop(index = filter.index, inplace = True)
print(cars_data)
```

```python
# Deleting Null Vlaues

cars_data.dropna(inplace = True)
print("Null vlaues Deleted")
```

```python
# Checking any duplicate values present in the dataset

duplicate_rows = cars_data.duplicated()
print(cars_data[duplicate_rows])
```

```python
# Deleting Duplicate Rows

cars_data.drop_duplicates(inplace = True)
print("Duplicate Rows Deleted")
```

```python
# Adding cleaned data into new dataset

cars_data.to_csv('cleaned_data.csv', index=False)
print("DataSaved")
```

```python
#2 Viewing datasets
```

```python
# Loading Datasets

final_data = pd.read_csv("cleaned_data.csv")
print(final_data)
```

```python
# Viewing number of rows and columns present in the dataset

print(final_data.shape)
```

```python
# viewing First five rows of the dataset

final_data.head()
```

```python
# Last five rows of the dataset

final_data.tail()
```

```python
# Viewing short summary of the datasets

print(final_data.info())
```

```python
# Viewing statistical description of the dataset

final_data.describe()
```

```python
# Counting number of cars present for each brand

print(final_data["Brand"].value_counts())
```

```python
#Removing km, commas and fullstops from the column to convert the column into integer datatype

final_data['kmDriven'] = final_data['kmDriven'].str.replace(' km', '', regex=False)  # Remove ' km'
final_data['kmDriven'] = final_data['kmDriven'].str.replace(',', '', regex=False)  # Remove commas
final_data['kmDriven'] = final_data['kmDriven'].str.replace('.', '', regex=False)  # Remove fullstops

# Converting column into integer
final_data["kmDriven"] = final_data["kmDriven"].astype("int64")
```

```python
#Removing ₹ and commas from the column to convert the column into integer datatype

final_data['AskPrice'] = final_data['AskPrice'].str.replace('₹', '', regex=False)
final_data['AskPrice'] = final_data['AskPrice'].str.replace(',', '', regex=False)  

# Converting column into integer
final_data["AskPrice"] = final_data["AskPrice"].astype("int64")
```

```python
# To check whether the column is converted into integert datatype or not 

final_data.info()
```

```python
#3 DATA VISUALIZATION
```

```python
# How Do Prices Vary for Different Owners Across Different Fuel Types?

# Boxplot of AskPrice by Owner and FuelType
plt.figure(figsize=(10, 6))
sns.boxplot(x='Owner', y='AskPrice', hue='FuelType', data=final_data)
plt.title('Price Distribution by Owner Count and Fuel Type')
plt.show()

```

```python
# Which brand has highest number of cars?

brand_counts = final_data['Brand'].value_counts()
brand_counts.plot(kind='bar', color='lightblue')
plt.title('Number of Cars by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Cars')
plt.show()
```

```python
# Which year has the highest ask price for cars?

plt.bar(final_data['Year'], final_data['AskPrice'], color = "red", alpha = 0.3)
plt.title('Year vs AskPrice')
plt.xlabel('Year')
plt.ylabel('Asking Price')
plt.show()

```

```python
# How many cars are available for each transmission type (Manual vs. Automatic)?

transmission_counts = final_data['Transmission'].value_counts()
transmission_counts.plot(kind='bar', color='lightgreen')
plt.title('Count of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Number of Cars')
plt.show()

print(transmission_counts)
```

```python
# What is the average price of cars by brand?

avg_price_by_brand = final_data.groupby('Brand')['AskPrice'].mean().sort_values()
plt.figure(figsize=(12, 6))
avg_price_by_brand.plot(kind='bar', color='red')
plt.title('Average Car Price by Brand')
plt.xlabel('Car Brand')
plt.ylabel('Average Ask Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# What is the average price of cars by fuel type (FuelType)?

import pandas as pd
import matplotlib.pyplot as plt
# Assuming you have the dataset as final_data
# Step 1: Calculate the average asking price by fuel type
avg_price_by_fuel = final_data.groupby('FuelType')['AskPrice'].mean().round(2)  # Round to 2 decimal places
# Step 2: Plot the results as a bar chart
plt.figure(figsize=(10, 6))
avg_price_by_fuel.plot(kind='bar', color='lightblue')
# Add title and labels
plt.title('Average Car Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Ask Price')
plt.xticks(rotation=0)
# Step 3: Display the average prices on top of the bars
for i, value in enumerate(avg_price_by_fuel):
    plt.text(i, value + 1000, f'{value:.2f}', ha='center', fontsize=10)
# Show the plot
plt.tight_layout()
plt.show()
# Print the rounded average prices for verification
```

```python
# Highest and lowest number of cars listed in which year?

year_counts = final_data['Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(year_counts.index, year_counts.values)
plt.title('Number of Cars Listed by Year')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()

print(year_counts)
```

```python
# Where Do the Most Expensive Cars Belong in Terms of Brand and Fuel Type?

import seaborn as sns
import matplotlib.pyplot as plt

# Create a pivot table for average prices by Brand and FuelType
pivot = final_data.pivot_table('AskPrice', ['Brand', 'FuelType'], aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(pivot, annot=True, cmap='YlGnBu', fmt='.1f', cbar_kws={'label': 'Average Price'})
plt.title('Average Price by Brand and Fuel Type')
plt.show()
```

```python
# How does the car's year correlate with asking price?

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='AskPrice', data=final_data)
plt.title('Car Year vs Asking Price')
plt.xlabel('Car Year')
plt.ylabel('Asking Price')
plt.show()

```

```python
# Conclusion:
# Overall, the asking price of a car is mainly influenced by factors like its age, kilometers driven, and brand. Newer cars with fewer kilometers tend to have higher prices,
# while older cars and those with higher mileage are generally priced lower. Additionally, the type of transmission and fuel also affect the price, with some variations
# depending on the brand and other features.
```
